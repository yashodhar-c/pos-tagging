```python
import pandas as pd
import nltk
df=pd.read_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews.csv")
```

```python
df.head()
```

```python
df=df[['review_title']]
df.head()
```

```python
pd.set_option('display.max_colwidth', None)
df.head()
```

```python
import re
import string
def punctuation_removal(txt):
    txt = txt.translate(str.maketrans('', '', string.punctuation))
    return txt

df['cleaned_text']=df['review_title'].apply(lambda title:punctuation_removal(title))
df
```

```python
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
```

```python
from string import digits
def remove_numbers(txt):
    txt=txt.translate(str.maketrans('','',digits))
    return txt

df['nonum_text']=df['cleaned_text'].apply(lambda title:remove_numbers(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python
def remove_special_characters(txt):
    txt=re.sub('[^ -~]',' ',txt)
    return txt

df['no_special_character_text']=df['nonum_text'].apply(lambda title:remove_special_characters(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python
df['review_title_lower']= df['no_special_character_text'].apply(lambda x: x.lower())
```

```python
from nltk.tokenize import word_tokenize
def tokenization_of_text(txt):
    tokens=word_tokenize(txt)
    return tokens
df['tokenized_text']=df['review_title_lower'].apply(lambda title:tokenization_of_text(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python
#nltk.download('stopwords')
from nltk.corpus import stopwords
def stop_word_removal(txt):
    stop_words=set(stopwords.words('english'))
    new_text=[title for title in txt if title not in stop_words]
    return new_text
df['stopword_removed_text']=df['tokenized_text'].apply(lambda title:stop_word_removal(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python
#nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

def lemmatized_text(txt):
    lemma_text=[lemmatizer.lemmatize(word) for word in txt]
    return lemma_text

df['lemmatized_text']=df['stopword_removed_text'].apply(lambda title:lemmatized_text(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python

def apply_pos_tags(txt):
    postag_text=nltk.pos_tag(txt)
    return postag_text
df['pos_tag_text']=df['lemmatized_text'].apply(lambda title:apply_pos_tags(title))
df.to_csv("C:\\Users\\Deepak J Bhat\\Downloads\\apple iphone reviews1.csv")
df.head()
```

```python
mylis=[]
mylis1=[]
for ind in df.index:
    if len(df['pos_tag_text'][ind])>=1:
        for x in df['pos_tag_text'][ind]:
            mylis.append(x[0])
            mylis1.append(x[1])
print(mylis)
print(mylis1)
print(len(mylis),len(mylis1))
```

```python
pos_tag_df=pd.DataFrame({"Words":mylis,"Pos_Tags":mylis1},columns=['Words','Pos_Tags'])
pos_tag_df
```

```python
pos_tag_df['Pos_Tags'].value_counts().nlargest(20)
```

```python
word_freq_dist=nltk.FreqDist(pos_tag_df['Pos_Tags'])
print(word_freq_dist)
```

```python
result=pd.DataFrame(word_freq_dist.most_common(20), columns=['Word','frequency'])
result['Percentage_Distribution']=((result['frequency'])/result['frequency'].sum())*100
result
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,5))
plt.bar(result['Word'], result['frequency'],color='orange',width=0.5)
plt.title("Word Frequency Visualization")
```

```python

from wordcloud import WordCloud
D={}
for ind in result.index:
    D[result['Word'][ind]]=result['frequency'][ind]
wordcloud = WordCloud(background_color ='black')
wordcloud.generate_from_frequencies(frequencies=D)
 
# plot the WordCloud image                       
plt.figure(figsize = (15, 15), facecolor = None)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
